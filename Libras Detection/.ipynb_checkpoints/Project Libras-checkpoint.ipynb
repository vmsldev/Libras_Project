```python
""" Importando as bibliotecas """
import mediapipe as mp
import cv2
import numpy as np
import os
import math
import csv
import pandas as pd
from sklearn import datasets
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
```

```python
os.mkdir('Imagens_Saida')
```

# Salvar Imagens da Mão

```python
cap = cv2.VideoCapture(0)
index = 0

def crop_img(img, p1, p2):
    """ Recorta um retângulo da imagem 'img' utilizando os
        pontos 'p1' e 'p2' como pontos delimitadores. """
    return img[p1[1]:p2[1], p1[0]:p2[0]]

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    while cap.isOpened():
        ret, frame = cap.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image = cv2.flip(image, 1)
        image = cv2.rectangle(image, (425,110),(640,430), (0,255,0), thickness=2, lineType=3, shift=0)
        image.flags.writeable = False
        results = hands.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        print(results)
        
             
        k = cv2.waitKey(1)
    
    
        if k % 256 == 32:
            cv2.imwrite(os.path.join('Imagens_Saida\Y', f'Y{index}.jpg'), crop_img(image, (425,110),(640,430)))
            index = index + 1
        
        cv2.imshow('Detectar mao', image)
    
        if k % 256 == 27:
            break
        
cap.release()
cv2.destroyAllWindows()
```

# Desenhar Esqueleto e Imprimir Distancias entre os Pontos

```python
image = cv2.imread(os.path.join('Imagens_Saida\I','I79.jpg'))

def distancia(ORIGx,ORIGy, x, y):

    distx = x-ORIGx

    disty = y-ORIGy

    valor = math.sqrt(distx*distx + disty*disty)

    return valor


def crop_img(image, p1, p2):
    """ Recorta um retângulo da imagem 'img' utilizando os
        pontos 'p1' e 'p2' como pontos delimitadores. """
    return img[p1[1]:p2[1], p1[0]:p2[0]]


with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    
        height, width, _ = image.shape
        
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.flip(image, 1)
        image = cv2.rectangle(image, (0,0),(0,0), (0,255,0), thickness=2, lineType=3, shift=0)
        image.flags.writeable = False
        results = hands.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        print(results)
        
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS,
                                         mp_drawing.DrawingSpec(color=(121,22,76),thickness=2, circle_radius=4),
                                         mp_drawing.DrawingSpec(color=(250,44,250),thickness=2, circle_radius=2),
                                         )
                DistDedao = distancia(hand.landmark[0].x, hand.landmark[0].y , hand.landmark[4].x, hand.landmark[4].y )
                DistInd = distancia(hand.landmark[0].x, hand.landmark[0].y , hand.landmark[8].x, hand.landmark[8].y )
                DistMeio = distancia(hand.landmark[0].x, hand.landmark[0].y , hand.landmark[12].x, hand.landmark[12].y )
                DistAnelar = distancia(hand.landmark[0].x, hand.landmark[0].y , hand.landmark[16].x, hand.landmark[16].y )
                Distdedin = distancia(hand.landmark[0].x, hand.landmark[0].y , hand.landmark[20].x, hand.landmark[20].y )
                
                print(DistDedao)
                print(DistInd)
                print(DistMeio)
                print(DistAnelar)
                print(Distdedin)
                
        
        cv2.imshow('Track de Mao.jpg', image)
        
        k = cv2.waitKey(0)
    

cv2.destroyAllWindows()
```

# Treinamento IA

## Inicio
* Dividindo os dados para treino e teste

```python
#Lendo o csv e guardando em uma variável
dfHands = pd.read_csv('letras_libras.csv', sep=',', encoding='utf-8')
```

```python
#Imprimindo o cabeçalho do dataset, para visualizar quais as colunas que o compõem
dfHands.head()
```

```python
#Analisando o tipo das variáveis para separar quais serão usados na regressão
dfHands.info()
```

```python
#Imprimindo as colunas do dataset
dfHands.columns
```

```python
#Salvando as colunas de rastreio na variável X (exceto a coluna não numérica)
X=dfHands[['DistDedao', 'DistInd', 'DistMeio', 'DistAnelar', 'Distdedin']]

#Salvando a coluna alvo na variável y
y=dfHands[['Letra']]
```

```python
#Separando os dados de treino e os dados para teste.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=110)
```

```python
#Teste de dados para o sensor
X_train[0:3]
```

```python
#Teste do resultado
y_train[0:3]
```

## Treino
* Utilizando Regressão Logística

```python
#Importando a função de uma biblioteca de modelos lineares
from sklearn.linear_model import LogisticRegression

#Modelo matematico usado para predição.
libras_model = LogisticRegression()
```

```python
#Realizando o treino das variáveis
libras_model.fit(X_train,y_train)
```

```python
#Variável para guardar a previsão
predito = libras_model.predict(X_test)
```

```python
#Imprimindo o resultado do teste realizado
y_test[0:20]
```

```python
#Confirmando a previsão
predito[0:20]
```

```python
#Pontuação do treino
libras_model.score(X_test,y_test)
```

## Avaliando o Modelo

```python
#Importando a função de métricas de classificação
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#Imprime um relatório de desempenho das predições, mostrando qual sua precisão, a média harmônica ponderada, o número de ocorrências reais, classificador de instancias positivas, e utilizando o 'unique()' para manter os dados agrupados
print(classification_report(y_test,predito,y_test.Letra.unique()))
```

## Treino v2

```python
#Importando a função de uma biblioteca de modelos sklearn
from sklearn import svm

#Modelo matematico usado para predição.
libras_svm = svm.SVC()
```

```python
#Variável para guardar a previsão
preditoSVM = libras_svm.predict(X_test)
```

```python
#Imprimindo o resultado do teste realizado
y_test[0:20]
```

```python
#Confirmando a previsão
preditoSVM[0:20]
```

```python
#Pontuação do treino
libras_svm.score(X_test,y_test)
```

# Utilizando o Modelo

```python
cap = cv2.VideoCapture(0)
alfabeto = cv2.imread('Alfabeto.png')

def distancia(ORIGx,ORIGy, x, y):

    distx = x-ORIGx

    disty = y-ORIGy

    valor = math.sqrt(distx*distx + disty*disty)

    return valor

def crop_img(img, p1, p2):
    """ Recorta um retângulo da imagem 'img' utilizando os
        pontos 'p1' e 'p2' como pontos delimitadores. """
    return img[p1[1]:p2[1], p1[0]:p2[0]]

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    while cap.isOpened():
        ret, frame = cap.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image = cv2.flip(image, 1)
        image = cv2.rectangle(image, (425,110),(640,430), (0,255,0), thickness=2, lineType=3, shift=0)
        image.flags.writeable = False
        results = hands.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        #print(results)
        
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS,
                                         mp_drawing.DrawingSpec(color=(121,22,76),thickness=2, circle_radius=4),
                                         mp_drawing.DrawingSpec(color=(250,44,250),thickness=2, circle_radius=2),
                                         )
                
                DistDedao = distancia(hand.landmark[0].x, hand.landmark[0].y , hand.landmark[4].x, hand.landmark[4].y )
                DistInd = distancia(hand.landmark[0].x, hand.landmark[0].y , hand.landmark[8].x, hand.landmark[8].y )
                DistMeio = distancia(hand.landmark[0].x, hand.landmark[0].y , hand.landmark[12].x, hand.landmark[12].y )
                DistAnelar = distancia(hand.landmark[0].x, hand.landmark[0].y , hand.landmark[16].x, hand.landmark[16].y )
                Distdedin = distancia(hand.landmark[0].x, hand.landmark[0].y , hand.landmark[20].x, hand.landmark[20].y )
                
            positions = [[DistDedao, DistInd, DistMeio, DistAnelar, Distdedin]]
                
            previsao = str(libras_svm.predict(positions))

            font = cv2.FONT_HERSHEY_PLAIN
            cv2.putText(image, previsao, (50, 50), font, 2, (0 , 191 , 255), 2)

             
        k = cv2.waitKey(1)
    
        cv2.imshow('Detectar Letra', image)
        cv2.imshow('Alfabeto', alfabeto)
    
        if k % 256 == 27:
            break
        
cap.release()
cv2.destroyAllWindows()
```
